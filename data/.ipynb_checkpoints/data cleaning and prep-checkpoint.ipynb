```python
def processFootball(filename):
    df = pd.read_csv(filename + '.csv')
    df = df[df['Score'].isna()]
    df = df[df['Home'].notna()]
    df['Date'] = pd.to_datetime(df['Date'])
    df['Name'] = df['Home'] + ' - ' + df['Away']
    df = df.drop(columns=['Wk', 'Day', 'Time', 'Score', 'Attendance', 'Venue', 'Referee', 'Match Report', 'Notes', 'Home', 'Away'])
    df.insert(0, 'Category', 'Football')
    #df.insert(0, 'ID', range(0, 0 + len(df)))
    df['Active'] = np.where(df['Date'] > pd.Timestamp.now(), True, False)
    #df.to_csv(filename + '_FINAL.csv')
    return df
```

```python
pt = processFootball('primeiraliga')
de = processFootball('bundesliga')
en = processFootball('epl')
es = processFootball('laliga')
fr = processFootball('ligue1')
it = processFootball('seriea')
```

```python
football = pd.concat([pt, de, en, es, fr, it])
football = football.drop(columns=['xG', 'xG.1'])
football.insert(0, 'ID', range(0, 0 + len(football)))
football
```

```python
football.to_csv('football.csv')
```
